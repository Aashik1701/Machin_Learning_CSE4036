## Simple Linear Regression



### Importing packages

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Create a DataFrame
data = {
    "Attendance": [85, 92, 78, 90, 88, 76, 95, 89, 83, 91, 80, 93, 87, 84, 94],
    "Grade": [88, 95, 79, 92, 90, 78, 97, 91, 85, 93, 82, 96, 89, 86, 95],
    "Performance": [86, 93, 80, 91, 89, 77, 96, 90, 84, 92, 81, 95, 88, 85, 94],
    "Extracurricular_Activity": [5, 8, 3, 7, 6, 2, 9, 6, 4, 8, 3, 9, 7, 5, 8]
}

df = pd.DataFrame(data)

print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Split the data into training and testing sets
X = df[["Attendance"]]
y = df["Grade"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Calculate the squared error and mean squared error
squared_error = (y_test - y_pred) ** 2
mean_squared_error = mean_squared_error(y_test, y_pred)

# Calculate the absolute error and mean absolute error
absolute_error = abs(y_test - y_pred)
mean_absolute_error = mean_absolute_error(y_test, y_pred)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)

print("Squared Error:", squared_error)
print("Mean Squared Error:", mean_squared_error)
print("Absolute Error:", absolute_error)
print("Mean Absolute Error:", mean_absolute_error)
print("R-squared:", r2)
```
